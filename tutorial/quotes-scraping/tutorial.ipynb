```python
from bs4 import BeautifulSoup
import requests
```

```python
link = 'http://quotes.toscrape.com'
html_text = requests.get(link).content
soup = BeautifulSoup(html_text, 'html.parser')
print(soup.prettify())
```

```python
# Scrape all Content in Boxes
content = soup.select('div.quote')
```

```python
# First Content Quote
content[0]
```

```python
# First Quote
first_quote = content[0].select_one('span.text').string
first_quote
```

```python
first_author = content[0].select_one('small.author').string
first_author
```

```python
first_tags = content[0].select('div.tags a.tag')
first_tags
```

```python
# Extract Tags
first_tags = [tag.string for tag in first_tags]
first_tags
```

```python
# Scrape Every Quote in Content
quotes = [quote.select_one('span.text').string for quote in content]
quotes
```

```python
# Scrape Every Author in Content
authors = [author.select_one('small.author').string for author in content]
authors
```
